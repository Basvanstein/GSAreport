```python
import pandas as pd
import numpy as np

mae = pd.read_csv('mse-all.csv', names=['label', 'function', 'dim', 'error', 'std'])  
```

```python
mae
```

```python
res = mae.groupby(['label', 'dim'])['error']

reformed_array = {}

for group_name, df_group in res:
    if (group_name[1] == 2):
        reformed_array[group_name[0]] = []
    score = df_group.mean() * 1000
    std = df_group.std() * 1000
    reformed_array[group_name[0]].append(f"${score:.3f}E^"+"{-3}$")# (±$"+f"{std:.2f}"+")E^{-3}$")
    #print(group_name, reformed_array[group_name[0]])

for alg in reformed_array:
    print (alg, " & \t", " & \t".join(reformed_array[alg]), "\\\\")
```
